```python
#Importing
import sklearn
from sklearn.datasets import load_wine
import pandas as pd 
import matplotlib.pyplot as plt

#loading
wine=load_wine()

#creating dataframe
column_name=wine.feature_names
Wdata=pd.DataFrame(wine.data)
Wdata.columns=column_name
Wdata.head(5)
```

```python
# FIRST METHOD of detecting outlier: box plot visualization

import seaborn as sns
bb=sns.boxplot(Wdata['magnesium'])
```

```python
outlier_data = bb.get_children()[6].get_data()
outlier_data
```

```python
import numpy as np
print(np.where(Wdata['magnesium']>138))
```

```python
# SECOND METHOD of detecting outlier: IQR method
Q1=np.percentile(Wdata['magnesium'],25,method='midpoint')
Q2=np.percentile(Wdata['magnesium'],50,method='midpoint')
Q3=np.percentile(Wdata['magnesium'],75,method='midpoint')

IQR=Q3-Q1
lower_limit=Q1-(1.5*IQR)
upper_limit=Q3+(1.5*IQR)

outlier=[]
for row in Wdata['magnesium']:
    if ((row<lower_limit) or (row>upper_limit)):
        outlier.append(row)
print("Outlier of the dataset:",outlier)
```
